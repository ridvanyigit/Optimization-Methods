# Golden Section Search Method

**Description:**

The Golden Section Search method is an optimization technique used to find the minimum or maximum value of single-variable, unimodal functions. This method progressively narrows the search interval to approach the optimal point. It is similar to the Fibonacci Search Method but utilizes the golden ratio constant instead of Fibonacci numbers. This ensures that the search interval is reduced by the same ratio in each iteration.

**Algorithm:**

1.  **Initialization:**
    *   Function to be optimized: $f(x)$
    *   Search interval: $[a, b]$
    *   Golden ratio constant: $\rho = \frac{3 - \sqrt{5}}{2} \approx 0.381966$
    *   Number of iterations: $N$ (or stopping criterion)

2.  **Point Calculation:**
    *   Determine two points within the search interval:
        *   $x_1 = a + \rho (b - a)$
        *   $x_2 = a + (1 - \rho) (b - a)$

3.  **Comparison and Interval Update:**
    *   If $f(x_1) > f(x_2)$: This indicates that the minimum point is more likely to be in the interval $[x_1, b]$. Therefore:
        *   $a = x_1$ (Update the left boundary of the search interval)
    *   Otherwise (if $f(x_1) \le f(x_2)$): The minimum point is more likely to be in the interval $[a, x_2]$. Therefore:
        *   $b = x_2$ (Update the right boundary of the search interval)

4.  **Stopping Criterion:**
    *   The algorithm stops if the specified number of iterations ($N$) has been reached or if a stopping criterion is met (e.g., the difference between function values falls below a tolerance value).
    *   Otherwise, return to step 2 and continue the iterations.

5.  **Result:**
    *   The final search interval $[a, b]$ is obtained. The midpoint of this interval is the closest estimate to the optimal point:
        *   $x_{opt} \approx \frac{a + b}{2}$

**Summary of Mathematical Formulation:**

\begin{align*}
\text{Golden Ratio Constant: } \rho &= \frac{3 - \sqrt{5}}{2} \\
\text{Calculation of Search Points: } \\
x_1 &= a + \rho (b - a) \\
x_2 &= a + (1 - \rho) (b - a) \\
\text{Interval Update Rules:} \\
\text{If } f(x_1) > f(x_2) \text{ then: } a &\leftarrow x_1 \\
\text{Otherwise: } b &\leftarrow x_2 \\
\text{Optimal Point Estimate: } x_{opt} &\approx \frac{a + b}{2}
\end{align*}

## 3. How It Works?

The Golden Section Search method works similarly to the Fibonacci Search Method, but it uses the golden ratio constant instead of Fibonacci numbers. In each iteration, two points ($x_1$ and $x_2$) within the interval are selected according to the golden ratio, and the function values at these points are compared. The interval containing the point with the higher function value is discarded, and the search interval is narrowed. This process continues for the specified number of iterations.

$$f(x) = x^4 - 14x^3 + 60x^2 - 70x$$
Iteration: 5

```python
# @title 1. Way
import numpy as np

def golden_section_search(f, a, b, max_iter=5):
    """Applies the golden section search algorithm and stores the intervals at each iteration."""
    phi = (1 + np.sqrt(5)) / 2  # Golden ratio
    intervals = []  # List to store the intervals

    for i in range(max_iter):
        c = b - (b - a) / phi
        d = a + (b - a) / phi

        if f(c) < f(d):
            b = d
        else:
            a = c

        intervals.append((a, b))

    return a, b, intervals

def objective_function(x):
    """Sample function to be optimized."""
    return x**4 - 14*x**3 + 60*x**2 - 70*x

a = -3  # Left boundary of the initial interval
b = 5   # Right boundary of the initial interval
max_iter = 5  # Maximum number of iterations

a_final, b_final, intervals = golden_section_search(objective_function, a, b, max_iter)

# Construct the Output String
output = "Golden Section Search Results:\n"
output += f"  a (final): {a_final:.6f}\n"
output += f"  b (final): {b_final:.6f}\n"
output += "\nIteration Intervals:\n"
for i, (a_val, b_val) in enumerate(intervals):
    output += f"  Iteration {i+1}: [{a_val:.6f}, {b_val:.6f}]\n"

# Estimated Minimum Point (Midpoint of the Interval)
minimum_x = (a_final + b_final) / 2
minimum_f = objective_function(minimum_x)

output += f"\nEstimated Minimum Point (x): {minimum_x:.6f}\n"
output += f"Estimated Minimum Value (f(x)): {minimum_f:.6f}\n"

print(output)
```

```python
# @title Visualization of `Interval Reduction` by Iteration
import matplotlib.pyplot as plt

# Create the plot
iteration_numbers = range(1, len(intervals) + 1)  # Iteration number based on the length of intervals
a_values = [interval[0] for interval in intervals]
b_values = [interval[1] for interval in intervals]

plt.figure(figsize=(10, 6))
plt.plot(iteration_numbers, a_values, marker='o', label='a (Left Boundary)')
plt.plot(iteration_numbers, b_values, marker='o', label='b (Right Boundary)')
plt.xlabel('Iteration Number')
plt.ylabel('Interval Boundaries')
plt.title('Golden Section Search - Interval Change by Iteration')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# @title 2. Way
import numpy as np
from scipy.optimize import minimize_scalar

# Function Definition
def objective_function(x):
    return x**4 - 14*x**3 + 60*x**2 - 70*x

# Initial Values
a = -3
b = 5

# Optimization with SciPy
result = minimize_scalar(objective_function,
                         bounds=(a, b),
                         method='bounded')

# Print Results
print("Optimization Results (SciPy):")
print(result)  # Prints all results
print("\nMinimum Value (x):", result.x)
print("Function Value (at minimum):", result.fun)
```

```python
# @title Optimization Results - `Function Graph & Minimum Point`
import numpy as np
import matplotlib.pyplot as plt

# Plotting the Function Graph
x = np.linspace(-5, 7, 400)  # Values for the x-axis (extended the range)
y = objective_function(x)      # Function values

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='f(x) = x^4 - 14x^3 + 60x^2 - 70x')

# Marking the Minimum Point
plt.plot(result.x, result.fun, 'ro', label='Minimum Point (SciPy)')  # Red dot

# Showing the Initial Intervals
plt.axvline(x=a, color='g', linestyle='--', label='Initial Interval (a)')
plt.axvline(x=b, color='g', linestyle='--', label='Initial Interval (b)')


# Labels and Title
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('SciPy minimize_scalar Results')
plt.grid(True)
plt.legend()
plt.show()
```
